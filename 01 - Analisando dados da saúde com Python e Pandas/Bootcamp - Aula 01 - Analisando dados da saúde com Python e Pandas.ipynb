### Desafio colocar uma coluna de região na tabela

```python
usaveis.index
```

```python
usaveis.index[0][0]
```

```python
usaveis.index[1][0]
```

```python
lista_regioes = []
for i in range(len(usaveis.index)):
    if usaveis.index[i][0] == '1':
        lista_regioes.append("Região Norte")
    elif usaveis.index[i][0] == '2':
        lista_regioes.append("Região Nordeste")
    elif usaveis.index[i][0] == '3':
        lista_regioes.append("Região Suldeste")
    elif usaveis.index[i][0] == '4':
        lista_regioes.append("Região Sul")
    elif usaveis.index[i][0] == '5':
        lista_regioes.append("Região Centro-oeste")
lista_regioes
```

```python
usaveis['Região'] = lista_regioes
usaveis
```

---

```python
usaveis.sort_values('Total', ascending=False)
```

```python
ordenado_por_total = usaveis.sort_values('Total', ascending=False)
ordenado_por_total = ordenado_por_total.drop("Total", axis=1)
ordenado_por_total = ordenado_por_total.drop("Região", axis=1)
ordenado_por_total.head()
```
